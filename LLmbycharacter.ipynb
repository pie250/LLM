```python
import torch
import torch.nn as nn
import torch.optim as optim

# 1. Sample training text
text = "hello world. this is a simple llm model. i wanna wants to say that you will definitely touch the target one day and you will achieve something which is dream of many."

chars = sorted(list(set(text)))
vocab_size = len(chars)

stoi = {ch: i for i, ch in enumerate(chars)}
itos = {i: ch for ch, i in stoi.items()}

# Encode text
data = torch.tensor([stoi[c] for c in text])

# 2. Model
class TinyLLM(nn.Module):
    def __init__(self, vocab_size, embed_size=32):
        super().__init__()
        self.embed = nn.Embedding(vocab_size, embed_size)
        self.fc = nn.Linear(embed_size, vocab_size)

    def forward(self, x):
        x = self.embed(x)
        x = self.fc(x)
        return x

model = TinyLLM(vocab_size)
loss_fn = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

# 3. Training
for epoch in range(500):
    optimizer.zero_grad()

    inputs = data[:-1]
    targets = data[1:]

    outputs = model(inputs)
    loss = loss_fn(outputs, targets)

    loss.backward()
    optimizer.step()

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}")

# 4. Text Generation
def generate(start="h", length=30):
    model.eval()
    idx = torch.tensor([stoi[start]])
    result = start

    for _ in range(length):
        with torch.no_grad():
            out = model(idx)
            probs = torch.softmax(out[-1], dim=0)
            next_idx = torch.multinomial(probs, 1)
            idx = torch.cat([idx, next_idx])
            result += itos[next_idx.item()]

    return result

print(generate("w"))
```
